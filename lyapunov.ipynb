```python
import numpy as np
import scipy as sp
from numpy import linalg as nla
import matplotlib.pyplot as plt
```

```python
def model(r):
    x=r[0]
    y=r[1]
    z=r[2]
    rho=0.3224
    delta=0.2625
    a=6.8927
    k=0.4032
    c=2.3952
    x1=1.0*(rho*delta*(x*x-a*x)+x*(x+y+c-c*np.tanh(x+z)))
    y1=-1.0*rho*delta*(a*y+x*x)
    z1=delta*(k-z-0.5*x)
    return (np.array([x1,y1,z1]))
```

```python
def jac(r,g1):
    p=0.3224
    d=0.2625
    a=6.8927
    x=r[0]
    y=r[1]
    M=np.array([[p*d*(2*x-a)+2*x+y,x,0],[-2*x*d*p,-p*x*d*p,0],[-d/2.0,0,-d]])
    res=np.matmul(M,g1)
    return(res)
```

```python
g1=np.array([1,0,0])
```

```python
tmax=800
h=0.01
t=0
x=-3
y=0
z=0
r=np.array([x,y,z])
talp=1
while (t<tmax):
    l1=jac(r,g1)
    l2=jac(r+0.5*h*l1,g1)
    l3=jac(r+0.5*h*l2,g1)
    l4=jac(r+h*l3,g1)
    
    g1=g1+(1.0/6)*h*(l1+2*l2+2*l3+l4)
    norm=np.sqrt(g1[0]**2+g1[1]**2+g1[2]**2)
    alp=np.log(norm)
    talp=talp+alp
    g1=g1/norm
    print(g1)
    
    k1=model(r)
    k2=model(r+0.5*h*k1)
    k3=model(r+0.5*h*k2)
    k4=model(r+k3*h)
    
    r=r+(1.0/6)*h*(k1+2*k2+2*k3+k4)
    t=t+h
exponent=1/(tmax)*talp
print(exponent)
```
